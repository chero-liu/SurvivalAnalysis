```python
import tarfile
import glob
import os
import json
import pandas as pd
```

```python
input = '/home/liuchenglong/Documents/TCGA/'
type = 'tpm_unstranded'
```

```python
def extract_tar_gz(file_path, extract_path='.'):
    """
    Extract a .tar.gz file to the specified extract path.

    :param file_path: str, path to the .tar.gz file.
    :param extract_path: str, directory to extract the .tar.gz file.
    """
    try:
        with tarfile.open(file_path, 'r:gz') as file:
            file.extractall(path=extract_path)
            print(f"File {file_path} extracted successfully to {extract_path}")
    except Exception as e:
        print(f"Error extracting {file_path}: {str(e)}")
```

```python
files = glob.glob(input+'/1.donwload/*')
os.makedirs(input+'/3.intergra',exist_ok=True)
```

```python
for file in files:
    os.makedirs(file.replace('1.donwload','2.untar')+'/gdc',exist_ok=True)
    os.makedirs(file.replace('1.donwload','2.untar')+'/clin',exist_ok=True)
    
    extract_tar_gz(glob.glob(file+'/gdc_download*')[0],file.replace('1.donwload','2.untar')+'/gdc')
    extract_tar_gz(glob.glob(file+'/clin*')[0],file.replace('1.donwload','2.untar')+'/clin')
    
    with open(glob.glob(file+"/metadata*.json")[0], 'r', encoding='utf-8') as file1:
    # 使用json.load()方法将文件中的JSON数据加载到一个Python对象中
        data = json.load(file1)
    raw_list=[]
    for i in range(len(data)):
        raw = pd.read_csv(input+"/2.untar/"+os.path.basename(file)+"/gdc/"+data[i]['file_id']+"/"+data[i]['file_name'],sep='\t',skiprows=1)
        raw = raw.drop([0,1,2,3])
        sub_raw = raw[['gene_id',type]]
        sub_raw = sub_raw.rename(columns={
            type: data[i]['associated_entities'][0]['entity_submitter_id']
        })
        raw_list.append(sub_raw)
    data = raw_list[0]
    for i in range(len(raw_list))[1:]:
        data = pd.merge(data,raw_list[i],how = 'inner',on='gene_id')
    data['gene_id'] = data['gene_id'].apply(lambda x: x.split('.')[0])
    data.to_csv(input+'/3.intergra/'+os.path.basename(file)+'_'+type+'.txt',index=None,sep='\t')
```
